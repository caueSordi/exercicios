## Exercicio 2



```python
initial_x = 0.1
n_vai = 5000

a_values = [1, 2, 3.8, 4]

variances = []
for a in a_values:
  # vai lá e pega os valores de a
  x = np.zeros(n_vai)
  x[0] = initial_x

  # faz todas as somatorias
  for i in range(1, n_vai):
    x[i] = a * x[i-1] * (1 - x[i-1])

  # calcula variação (GRAÇAS A DEUS TEM UMA FUNÇÃO PRA ISSO)
  variance = np.var(x)
  print(f"variação para a = {a}:", variance)
  # salva a variação (pra depois fazer a média
  variances.append(variance)
  

# Imprime vai lá e imprime os valores
variacao_media = np.median(variances)
print("variação média:")
print(variacao_media)

```
