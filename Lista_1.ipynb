```python
import numpy as np
import matplotlib.pyplot as plt

# Valores iniciais
initial_x = 0.1
n_vai = 5000

# Valores A
a_values = [1, 2, 3.8, 4]

# Variações
variances = []

# Iterar sobre os valores de a
for a in a_values:
  # Inicializar x
  x = np.zeros(n_vai)
  x[0] = initial_x

  # Calcular os valores de x
  for i in range(1, n_vai):
    x[i] = a * x[i-1] * (1 - x[i-1])

  # faz a variação
  variance = np.var(x)

  #varianca
  variances.append(variance)

  print(f"Variação para a = {a}: {variance}")

# Plotar variações em função de a_values
plt.plot(a_values, variances, linestyle='-')

# Adicionar título e rótulos
plt.title("Variação em função de a")
plt.xlabel("valores de a")
plt.ylabel("Variação")

# Mostrar o gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Classe que representa uma aresta
class Aresta:
    def __init__(self, no1, no2):
        self.no1 = no1
        self.no2 = no2

# Classe que faz uma rede
class Network:
    def __init__(self, num_nos):
        self.num_nos = num_nos
        self.lista_adj = [[] for _ in range(num_nos)]
        self.coordenadas = np.zeros((num_nos, 2))

    def adicionar_aresta(self, no1, no2):
        self.lista_adj[no1].append(no2)
        self.lista_adj[no2].append(no1)

    def adicionar_coordenadas_no(self, no, x, y):
        self.coordenadas[no] = [x, y]

    def plotar_rede(self):
        for i in range(self.num_nos):
            for j in self.lista_adj[i]:
                plt.plot([self.coordenadas[i][0], self.coordenadas[j][0]],
                         [self.coordenadas[i][1], self.coordenadas[j][1]], 'b-')
        plt.plot(self.coordenadas[:, 0], self.coordenadas[:, 1], 'ro')
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.title('Rede')
        plt.grid(True)
        plt.show()

    def deletar_no(self, no):
        # Deletar arestas conectadas ao no 
        for vizinho in self.lista_adj[no]:
            self.lista_adj[vizinho].remove(no)
        del self.lista_adj[no]

    def deletar_aresta(self, no1, no2):
        self.lista_adj[no1].remove(no2)
        self.lista_adj[no2].remove(no1)
        if not self.lista_adj[no1]:
            del self.lista_adj[no1]
        if not self.lista_adj[no2]:
            del self.lista_adj[no2]

    def adicionar_aresta_entre_nos(self, no1, no2):
        self.adicionar_aresta(no1, no2)

# Exemplo
rede = Network(5)
rede.adicionar_coordenadas_no(0, 0, 0)
rede.adicionar_coordenadas_no(1, 0, 1)
rede.adicionar_coordenadas_no(2, 2, 2)
rede.adicionar_coordenadas_no(3, 3, 1)
rede.adicionar_coordenadas_no(4, 2, 0)
rede.adicionar_aresta(0, 1)
rede.adicionar_aresta(1, 2)
rede.adicionar_aresta(2, 3)
rede.adicionar_aresta(3, 4)
rede.adicionar_aresta(4, 0)

rede.plotar_rede()

# Deletar um no e arestas
rede.deletar_no(2)
rede.plotar_rede()

# Deletararesta
rede.deletar_aresta(1, 3)
rede.plotar_rede()

# Adicionar uma aresta entre nós existentes
rede.adicionar_aresta_entre_nos(0, 3)
rede.plotar_rede()
```
