```python
import numpy as np
import matplotlib.pyplot as plt

# Valores iniciais
initial_x = 0.1
n_vai = 5000

# Valores A
a_values = [1, 2, 3.8, 4]

# Variações
variances = []

# Iterar sobre os valores de a
for a in a_values:
  # Inicializar x
  x = np.zeros(n_vai)
  x[0] = initial_x

  # Calcular os valores de x
  for i in range(1, n_vai):
    x[i] = a * x[i-1] * (1 - x[i-1])

  # faz a variação
  variance = np.var(x)

  #varianca
  variances.append(variance)

  print(f"Variação para a (numpy)= {a}: {variance}")

# Plotar variações em função de a_values
plt.plot(a_values, variances, linestyle='-')

# Adicionar título e rótulos
plt.title("Variação em função de a  ")
plt.xlabel("valores de a")
plt.ylabel("Variação")
plt.grid(True)

# Mostrar o gráfico
plt.show()

# Clear the 'variances' list before manual calculation
variances.clear()

#manualmente
for a in a_values:
    x = np.zeros(n_vai)
    x[0] = initial_x

    for i in range(1, n_vai):
        x[i] = a * x[i-1] * (1 - x[i-1])

    mean = sum(x) / len(x)
    varian = sum((xi - mean) ** 2 for xi in x) / len(x)
    variances.append(varian)

print("variação manual para (A):")
for i in range(len(a_values)):
    print(f"a = {a_values[i]}: {variances[i]}")

plt.plot(a_values, variances)
plt.title("Variation of X for each a")
plt.xlabel("a")
plt.ylabel("Variance of X")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Valores iniciais
initial_x = 0.1
n_vai = 5000

# Valores de A
a_values = [1, 2, 3.8, 4]

# Variações
variances = []
means = []

# Iterar sobre os valores de A
for a in a_values:
    # Inicializar x
    x = np.zeros(n_vai)
    x[0] = initial_x

    # Calcular os valores de x
    for i in range(1, n_vai):
        x[i] = a * x[i-1] * (1 - x[i-1])

    # Calcula a média e a variância
    mean = np.mean(x)
    variance = np.var(x)

    # Adiciona média e variância às respectivas listas
    means.append(mean)
    variances.append(variance)

# Plotar variações em função de A
plt.figure(figsize=(12, 8))

# Média Manual
plt.subplot(2, 2, 1)
plt.plot(a_values, means, marker='o')
plt.title('Média Manual')
plt.xlabel('Valores de A')
plt.ylabel('Média')
plt.grid(True)

# Variância Manual
plt.subplot(2, 2, 2)
plt.plot(a_values, variances, marker='o')
plt.title('Variância Manual')
plt.xlabel('Valores de A')
plt.ylabel('Variância')
plt.grid(True)

# Cálculos com NumPy
np_variances = []
np_means = []

for a in a_values:
    # Inicializar x
    x = np.zeros(n_vai)
    x[0] = initial_x

    # Calcular os valores de x
    for i in range(1, n_vai):
        x[i] = a * x[i-1] * (1 - x[i-1])

    # Calcula a média e a variância com NumPy
    np_mean = np.mean(x)
    np_variance = np.var(x)

    # Adiciona média e variância às respectivas listas
    np_means.append(np_mean)
    np_variances.append(np_variance)

# Média com NumPy
plt.subplot(2, 2, 3)
plt.plot(a_values, np_means, marker='o')
plt.title('Média com NumPy')
plt.xlabel('Valores de A')
plt.ylabel('Média')
plt.grid(True)

# Variância com NumPy
plt.subplot(2, 2, 4)
plt.plot(a_values, np_variances, marker='o')
plt.title('Variância com NumPy')
plt.xlabel('Valores de A')
plt.ylabel('Variância')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Exercicio 4

```python
import numpy as np
import matplotlib.pyplot as plt

# Valores iniciais
X = []
initial_x = 0.6
r_min = 1
r_max = 4
dr = 0.001
n = int((r_max - r_min) / dr)

# Iterar sobre os valores de r
for r in np.arange(r_min, r_max, dr):
    x = [initial_x]
    for i in range(1, 5000):
        next_x = r * x[-1] * (1 - x[-1])
        x.append(next_x)
    X.append(x[100:])


# pega os valores de r e x e plota
plt.figure(figsize=(10, 6))
for i in range(n):
    r = r_min + dr * i
    x = X[i]
    plt.plot([r] * len(x), x, '.', markersize=0.01, color='blue')

plt.xlabel('A')
plt.ylabel('X')
plt.title('diagrama de bifurcação')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# ----------------- Aresta -----------------
class Aresta:
    def __init__(self, no1, no2):  
        self._No1 = no1
        self._No2 = no2

    def NosConectados(self):
        return [self._No1, self._No2]

# ---------------- classe No ----------------
class No:
    def __init__(self, coord, info):
        self._Arestas = []
        self._Coord = coord
        self._Info = info

    def con_no(self, no):
        for aresta in self._Arestas:
            if aresta.NosConectados()[1] == no:
                return
        
        self._Arestas.append(Aresta(self, no))
        no._Arestas.append(Aresta(no, self))
    
    def DesconectarNo(self, no, nos):
        for aresta in self._Arestas:
            if aresta.NosConectados()[1] == no:
                self._Arestas.remove(aresta)
                break

        for aresta in no._Arestas:
            if aresta.NosConectados()[1] == self:
                no._Arestas.remove(aresta)
                break
                
        if not no._Arestas:
            nos.remove(no)
            del no

        if not self._Arestas:
            nos.remove(self)
            del self

    def DeletarNo(self, nos):
        for aresta in reversed(self._Arestas):
            aresta.NosConectados()[1].DesconectarNo(self, nos)

    def ArestasConectadas(self):
        return self._Arestas

    def Coord(self):
        return "X: " + str(self._Coord[0]) + "\tY: " + str(self._Coord[1])

    def Info(self):
        return self._Info

# posição do no
def No_pos(info, nos):
    for no in nos:
        if(no.Info() == info):
            return nos.index(no)


# Criando uma lista vazia para armazenar os nós
nos = []

# Criando 10 nós e adicionando-os à lista
for i in range(10):
    nos.append(No([i, i*2], i))

# -----------------Conexões-----------------
    
nos[7].con_no(nos[5])

nos[7].con_no(nos[6])

nos[7].con_no(nos[8])

nos[1].con_no(nos[2])

nos[1].con_no(nos[3])

nos[2].con_no(nos[4])

nos[2].con_no(nos[6])

nos[3].con_no(nos[1])

nos[3].con_no(nos[4])

nos[4].con_no(nos[2])

nos[0].con_no(nos[1])

nos[4].con_no(nos[3])

nos[5].con_no(nos[6])

nos[5].con_no(nos[7])

nos[6].con_no(nos[2])

nos[6].con_no(nos[5])

nos[6].con_no(nos[7])

nos[8].con_no(nos[7])

nos[8].con_no(nos[9])

nos[9].con_no(nos[8])

# Variável para controlar o encerramento do programa
end = False
while not end:
    conexoes = []
    for no in nos:
        if no.ArestasConectadas():
            for aresta in no.ArestasConectadas():
                conexoes.append([aresta.NosConectados()[0].Info(), aresta.NosConectados()[1].Info()])

    # ------------- Criando graph do networkx -------------
    grafo = nx.Graph()

    grafo.add_nodes_from([no.Info() for no in nos])
    grafo.add_edges_from(conexoes)

    # desenhando o grafo
    nx.draw(grafo, with_labels=True, font_weight='bold')
    plt.show()        

# ------------ comandos ------------
def menu():
    print("\nComandos:")
    print("conectar no1 , no2- Conectar dois nós")
    print("desconectar no1 , no 2- Deletar conexão")
    print("deletar no - Deletar nó")
    print("fim - Encerrar")

    # ---------------menu----------------
    menu()
    
    # ler input
    escolha = input(" ").strip()

    if escolha == "fim":
        end = True

    elif escolha == "conectar":
        no1 = No_pos(int(input("1° nó: ")), nos)
        no2 = No_pos(int(input("2° nó: ")), nos)
        nos[no1].con_no(nos[no2])

    elif escolha == "desconectar":
        no1 = No_pos(int(input("1° nó: ")), nos)
        no2 = No_pos(int(input("2° nó: ")), nos)
        nos[no1].DesconectarNo(nos[no2], nos)
        
    elif escolha == "deletar":
        no = No_pos(int(input("nó deletado: ")), nos)
        nos[no].DeletarNo(nos)
```
