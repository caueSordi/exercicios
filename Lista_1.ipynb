```python
import numpy as np
import matplotlib.pyplot as plt

# Valores iniciais
initial_x = 0.1
n_vai = 5000

# Valores A
a_values = [1, 2, 3.8, 4]

# Variações
variances = []

# Iterar sobre os valores de a
for a in a_values:
  # Inicializar x
  x = np.zeros(n_vai)
  x[0] = initial_x

  # Calcular os valores de x
  for i in range(1, n_vai):
    x[i] = a * x[i-1] * (1 - x[i-1])

  # faz a variação
  variance = np.var(x)

  #varianca
  variances.append(variance)

  print(f"Variação para a (numpy)= {a}: {variance}")

# Plotar variações em função de a_values
plt.plot(a_values, variances, linestyle='-')

# Adicionar título e rótulos
plt.title("Variação em função de a  ")
plt.xlabel("valores de a")
plt.ylabel("Variação")
plt.grid(True)

# Mostrar o gráfico
plt.show()

# Clear the 'variances' list before manual calculation
variances.clear()

#manualmente
for a in a_values:
    x = np.zeros(n_vai)
    x[0] = initial_x

    for i in range(1, n_vai):
        x[i] = a * x[i-1] * (1 - x[i-1])

    mean = sum(x) / len(x)
    varian = sum((xi - mean) ** 2 for xi in x) / len(x)
    variances.append(varian)

print("variação manual para (A):")
for i in range(len(a_values)):
    print(f"a = {a_values[i]}: {variances[i]}")

plt.plot(a_values, variances)
plt.title("Variation of X for each a")
plt.xlabel("a")
plt.ylabel("Variance of X")
plt.grid(True)
plt.show()
```

## Exercicio 4

```python
import numpy as np
import matplotlib.pyplot as plt

# Valores iniciais
X = []
initial_x = 0.6
r_min = 1
r_max = 4
dr = 0.001
n = int((r_max - r_min) / dr)

# Iterar sobre os valores de r
for r in np.arange(r_min, r_max, dr):
    x = [initial_x]
    for i in range(1, 5000):
        next_x = r * x[-1] * (1 - x[-1])
        x.append(next_x)
    X.append(x[100:])


# pega os valores de r e x e plota
plt.figure(figsize=(10, 6))
for i in range(n):
    r = r_min + dr * i
    x = X[i]
    plt.plot([r] * len(x), x, '.', markersize=0.01, color='blue')

plt.xlabel('A')
plt.ylabel('X')
plt.title('diagrama de bifurcação')
plt.show()
```

```python
class Aresta:
    def __init__(self, no1, no2):  
        self._No1 = no1
        self._No2 = no2

    def NosConectados(self):
        return [self._No1, self._No2]

class No:
    def __init__(self, coord):
        self.Arestas = []
        self.Coord = coord

    def ConectarNo(self, no):
        for aresta in self.Arestas:
            if aresta.NosConectados()[1] == no:
                print("Conexão repetida!")
                return
        
        self.Arestas.append(Aresta(self, no))
        no.Arestas.append(Aresta(no, self))
    
    def DesconectarNo(self, no, nos):
        for aresta in self.Arestas:
            if aresta.NosConectados()[1] == no:
                self.Arestas.remove(aresta)
                break

        for aresta in no.Arestas:
            if aresta.NosConectados()[1] == self:
                no.Arestas.remove(aresta)
                break
                
        if len(no.Arestas) == 0:
            nos.remove(no)

        if len(self.Arestas) == 0:
            nos.remove(self)

    def DeletarNo(self, nos):
        #Ele opera de forma invertida pois, ao apagar a conexão, ele pula indíces da lista
        for aresta in reversed(self.Arestas):
            aresta.NosConectados()[1].DesconectarNo(self, nos)

    def ArestasConectadas(self):
        return self.Arestas

    def Info(self):
        return "X: " + str(self.Coord[0]) + "\tY: " + str(self.Coord[1])

nos = []

for i in range(10):
    nos.append(No([i, i*2]))

#No 0
nos[0].ConectarNo(nos[1])

#No 1
nos[1].ConectarNo(nos[2])
nos[1].ConectarNo(nos[3])

#No 2
nos[2].ConectarNo(nos[4])
nos[2].ConectarNo(nos[6])

#No 3
nos[3].ConectarNo(nos[1])
nos[3].ConectarNo(nos[4])

#No 4
nos[4].ConectarNo(nos[2])
nos[4].ConectarNo(nos[3])

#No 5
nos[5].ConectarNo(nos[6])
nos[5].ConectarNo(nos[7])

#No 6
nos[6].ConectarNo(nos[2])
nos[6].ConectarNo(nos[5])
nos[6].ConectarNo(nos[7])

#No 7
nos[7].ConectarNo(nos[5])
nos[7].ConectarNo(nos[6])
nos[7].ConectarNo(nos[8])

#No 8
nos[8].ConectarNo(nos[7])
nos[8].ConectarNo(nos[9])

#No 9
nos[9].ConectarNo(nos[8])

fim = 0

while fim == 0:    
    print("\n\n")
    for no in nos:
        print("Posição na lista: ", nos.index(no), ": ", no.Info(), " Conexões: ", end='')
        for aresta in no.ArestasConectadas():
            print(nos.index(aresta.NosConectados()[1]), end = ' ')
        print("\n")
        
    
    escolha = int(input("\nComandos:\n1 - Conectar dois nós\n2 - Deletar conexão\n3 - Deletar nó\n4 - Encerra\n"))
    if(escolha == 4):
        fim = 1
    if(escolha == 1):
        no1 = int(input("Primeiro nó: "))
        no2 = int(input("Segundo nó: "))
        nos[no1].ConectarNo(nos[no2])
    if(escolha == 2):
        no1 = int(input("Primeiro nó: "))
        no2 = int(input("Segundo nó: "))
        nos[no1].DesconectarNo(nos[no2], nos)
    if(escolha == 3):
        no1 = int(input("Qual nó: "))
        nos[no1].DeletarNo(nos)
        
```
