```python
import numpy as np
import matplotlib.pyplot as plt

# Valores iniciais
initial_x = 0.1
n_vai = 5000

# Valores A
a_values = [1, 2, 3.8, 4]

# Variaes
variances = []

# Iterar sobre os valores de a
for a in a_values:
  # Inicializar x
  x = np.zeros(n_vai)
  x[0] = initial_x

  # Calcular os valores de x
  for i in range(1, n_vai):
    x[i] = a * x[i-1] * (1 - x[i-1])

  # faz a variao
  variance = np.var(x)

  #salva
  variances.append(variance)

# Plotar variaes em funo de a_values
plt.plot(a_values, variances, 'o-', linestyle='-')

# Adicionar ttulo e rtulos
plt.title("Variao em funo de a")
plt.xlabel("valores de a")
plt.ylabel("Variao")

# Mostrar o grfico
plt.show()
```

```python
initial_x = 0.1
n_vai = 5000

a_values = [1, 2, 3.8, 4]

variances = []
for a in a_values:
  # vai l e pega os valores de a
  x = np.zeros(n_vai)
  x[0] = initial_x

  # faz todas as somatorias
  for i in range(1, n_vai):
    x[i] = a * x[i-1] * (1 - x[i-1])
    #print(x[i])

  # calcula variao (GRAAS A DEUS TEM UMA FUNO PRA ISSO)
  variance = np.var(x)
  print(f"variao para a = {a}:", variance)
  # salva a variao (pra depois fazer a mdia
  variances.append(variance)
  

# Imprime vai l e imprime os valores
variacao_media = np.median(variances)
print("variao mdia:")
print(variacao_media)

```
